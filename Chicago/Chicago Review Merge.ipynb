```python
import pandas as pd
import numpy as np
df0_5 = pd.read_csv('chicago0_5.csv')
df1 = pd.read_csv('chicago1.csv')
df2 = pd.read_csv('chicago2.csv')
df3 = pd.read_csv('chicago3.csv')
df4 = pd.read_csv('chicago4.csv')
df5 = pd.read_csv('chicago5.csv')
df = pd.concat([df0_5, df1, df2, df3, df4, df5], axis = 0)
```

```python
df.drop_duplicates(subset=['Review'])
```

```python
restaurants = df['Restaurant Name'].unique()
dff_list = []
for rest in restaurants:
    dff = df[df['Restaurant Name'] == rest]
    dff['Rating2'] = dff.Rating.expanding().mean().round(1)
    dff['rating_seen'] = dff['Rating2'].shift(1)
    dff = dff.reset_index(drop=True)
    dff_list.append(dff)
final_df = pd.concat(dff_list)
```
