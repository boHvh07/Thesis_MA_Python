```python
import pandas as pd
import numpy as np
df1 = pd.read_csv('la1.csv')
df2 = pd.read_csv('la2.csv')
df3 = pd.read_csv('la3.csv')
df4 = pd.read_csv('la4.csv')
df5 = pd.read_csv('la5.csv')
df6 = pd.read_csv('la6.csv')
df7 = pd.read_csv('la7.csv')
df8 = pd.read_csv('la8.csv')
df9 = pd.read_csv('la9.csv')
df10 = pd.read_csv('la205tillend.csv')
df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10], axis = 0)
```

```python
df.drop_duplicates(subset=['Review'])
```

```python
restaurants = df['Restaurant Name'].unique()
dff_list = []
for rest in restaurants:
    dff = df[df['Restaurant Name'] == rest]
    dff['Rating2'] = dff.Rating.expanding().mean().round(1)
    dff['rating_seen'] = dff['Rating2'].shift(1)
    dff = dff.reset_index(drop=True)
    dff_list.append(dff)
final_df = pd.concat(dff_list)
```
